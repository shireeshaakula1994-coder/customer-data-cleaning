```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# -----------------------------
# Step 0: Create Data
# -----------------------------
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age':         [25, np.nan, 35, np.nan, np.nan, 40],
    'city':        ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender':      ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

# -----------------------------
# Task 1: Clean the Data
# -----------------------------
# Fill missing values
df['age'] = df['age'].fillna(df['age'].median())
df['city'] = df['city'].fillna(df['city'].mode()[0])

# Remove duplicates
df = df.drop_duplicates()
print("\nAfter cleaning:\n",df)

# -----------------------------
# Task 2: Encoding
# -----------------------------
# One-Hot Encoding for city
df = pd.get_dummies(df, columns=['city'])

# Label Encoding for gender
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})

print("After Encoding:\n", df)

# -----------------------------
# Task 3: Scaling
# -----------------------------
columns = ['age', 'annual_income']

# Min-Max Scaling
minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[columns] = minmax.fit_transform(df_minmax[columns])

print("\nMin-Max Scaled Data:\n", df_minmax)

# Standardization
standard = StandardScaler()
df_standard = df.copy()
df_standard[columns] = standard.fit_transform(df_standard[columns])

print("\nStandardized Data:\n", df_standard)
```
